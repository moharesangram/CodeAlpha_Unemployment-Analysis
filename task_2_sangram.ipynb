```python
# Rename columns for simplicity
df.columns = [
    "Region",
    "Date",
    "Frequency",
    "Estimated Unemployment Rate (%)",
    "Estimated Employed",
    "Estimated Labour Participation Rate (%)",
    "Area"
]

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Drop missing values
df.dropna(inplace=True)

df.head()
```

```python
df.describe()
```

```python
region_avg = df.groupby("Region")["Estimated Unemployment Rate (%)"].mean().sort_values()

plt.figure()
region_avg.plot(kind="barh")
plt.xlabel("Average Unemployment Rate (%)")
plt.title("Average Unemployment Rate by Region")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Unemployment in India.csv")
df.columns = df.columns.str.strip()

```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(
    x="Area",
    y="Estimated Unemployment Rate (%)",
    data=df,
    order=["Rural", "Urban"]
)
plt.title("Unemployment Rate: Urban vs Rural")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xlabel("Area")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df = df.groupby("Date", as_index=False)["Estimated Unemployment Rate (%)"].mean()
```

```python
before_covid = df[df["Date"] < "2020-03-01"]
after_covid = df[df["Date"] >= "2020-03-01"]

plt.figure()
plt.plot(before_covid["Date"], before_covid["Estimated Unemployment Rate (%)"], label="Before COVID")
plt.plot(after_covid["Date"], after_covid["Estimated Unemployment Rate (%)"], label="After COVID")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.title("Impact of COVID-19 on Unemployment Rate")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.lineplot(x="Date", y="Estimated Labour Participation Rate (%)", data=df)
plt.title("Labour Participation Rate Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.heatmap(
    df[[
        "Estimated Unemployment Rate (%)",
        "Estimated Employed",
        "Estimated Labour Participation Rate (%)"
    ]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Matrix")
plt.show()

```
